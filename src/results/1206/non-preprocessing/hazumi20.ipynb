```python
import yaml 
import pickle
import glob 
import pandas as pd
files = glob.glob("*.yaml")
print(f"実験結果数 : {len(files)}")

version = "2010"
with open(f'../../data/Hazumi_features/Hazumi{version}_features.pkl', mode='rb') as f:
    SS, SS_binary, SS_ternary, TS, TS_binary, TS_ternary, SP, TP, Text, Audio, Visual, vid \
        = pickle.load(f, encoding='utf-8')

def profiling(profile, id, TP=None):
    age = int(id[5])
    gender = id[4]

    age_th = 3

    if profile == 0:
        res = 0
    elif profile == 1:
        # 性別
        if gender == 'F':
            res = 0
        else:
            res = 1
    elif profile == 2:
        # 年齢(2クラス)
        # 40 <=, 40 >
        if age <= age_th:
            res = 0
        else:
            res = 1
    elif profile == 3:
        if gender == 'F':
            if age <= age_th:
                res = 0 
            else:
                res = 1
        else:
            if age <= age_th:
                res = 2
            else:
                res = 3
    elif profile == 4:
        res = 1 if TP[id][0] >= 8 else 0
    elif profile == 5:
        res = 1 if TP[id][1] >= 8 else 0
    elif profile == 6:
        res = 1 if TP[id][2] >= 8 else 0
    elif profile == 7:
        res = 1 if TP[id][3] >= 8 else 0
    elif profile == 8:
        res = 1 if TP[id][4] >= 8 else 0
    return res
```

# TS & ternary

```python
df_results = pd.DataFrame(0, index=[0,1,2,3,4,5,6,7,8], columns=['t','a','v','ta','tv','av','tav'])
for file in files:
    with open(file) as file:
        obj = yaml.safe_load(file)
        config = obj["config"]
        if config["annot"] == "TS" and config["label"] == "ternary":
            results = obj["results"]
            data = list(results.values())
            acc = round(sum(data) / len(data), 3)
            df_results.at[config['pmode'], config['modal']] = acc

display(df_results)
```

### 性別

```python
# 女性
df_results = pd.DataFrame(0, index=[0,1,2,3,4,5,6,7,8], columns=['t','a','v','ta','tv','av','tav'])
for file in files:
    with open(file) as file:
        obj = yaml.safe_load(file)
        config = obj["config"]
        if config["annot"] == "TS" and config["label"] == "ternary":
            # results = obj["results"]
            results = []
            for key, data in obj["results"].items():
                if profiling(1, key) == 0:
                    results.append(data)
            acc = round(sum(results) / len(results), 3)
            df_results.at[config['pmode'], config['modal']] = acc
print("女性")
display(df_results)

# 男性
df_results = pd.DataFrame(0, index=[0,1,2,3,4,5,6,7,8], columns=['t','a','v','ta','tv','av','tav'])
for file in files:
    with open(file) as file:
        obj = yaml.safe_load(file)
        config = obj["config"]
        if config["annot"] == "TS" and config["label"] == "ternary":
            # results = obj["results"]
            results = []
            for key, data in obj["results"].items():
                if profiling(1, key) == 1:
                    results.append(data)
            acc = round(sum(results) / len(results), 3)
            df_results.at[config['pmode'], config['modal']] = acc
print("男性")
display(df_results)
```
