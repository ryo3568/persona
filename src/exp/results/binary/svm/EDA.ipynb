```python
import yaml 
import glob 
import pandas as pd
pd.options.display.precision = 3


label = "binary"
model = "svm"

results_files = sorted(glob.glob(f"*.yaml"))
results_len = len(results_files)
results = {}

for i, results_file in enumerate(results_files):
    with open(results_file) as file:
        result = yaml.safe_load(file)
        results[i] = result
```

```python
# 性別ごとの結果
for i in range(results_len):
    if i != 0 and i != 1:
        continue
    columns = ["# users", "Acc", "F1"]
    index = ["全員", "女性", "男性"]
    All = [0, 0, 0]
    Female = [0, 0, 0]
    Male = [0, 0, 0]
    result = results[i]
    for user in result.keys():
        if user == "all":
            continue
        age = int(user[5])
        gender = user[4]
        if gender == 'F':
            Female[0] += 1
            Female[1] += result[user]["acc"]
            Female[2] += result[user]["f1"]
        else:
            Male[0] += 1
            Male[1] += result[user]["acc"]
            Male[2] += result[user]["f1"]
        All[0] += 1
        All[1] += result[user]["acc"]
        All[2] += result[user]["f1"]
    print(f"profile {i} : ")
    df = pd.DataFrame(data=[All, Female, Male], index=index, columns=columns)
    df.loc[:, "Acc"] /= df.loc[:, "# users"]
    df.loc[:, "F1"] /= df.loc[:, "# users"]
    print(df)
    print("=" * 30)
```
